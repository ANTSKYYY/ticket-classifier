```python
import pandas as pd

# Remplace par le nom exact de ton fichier d'entraînement AVANT ta fusion
chemin_csv = 'train_augmented.csv' 

try:
    df = pd.read_csv(chemin_csv)
    print("🕵️‍♂️ AUDIT DU DATASET D'ENTRAÎNEMENT\n" + "="*40)
    
    # On vérifie chaque classe existante dans le fichier
    for label in df['label'].unique():
        if pd.isna(label): continue
        print(f"\n🏷️ CLASSE DANS LE FICHIER : {label}")
        print("-" * 40)
        
        # On prend 4 phrases au hasard pour cette classe
        echantillons = df[df['label'] == label]['text'].dropna().sample(4).values
        for texte in echantillons:
            print(f"💬 {texte}")
            
except Exception as e:
    print(f"❌ Erreur lors de la lecture : {e}")
```
